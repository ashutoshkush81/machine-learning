```python
# How to draw a pdf(Probability distribution function.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
iris = pd.read_csv('iris.csv')
count,bins_edges = np.histogram(iris['petal.length'],bins=10,density=True)
pdf = count/(sum(count))
print(pdf)
print(bins_edges)
```

```python
cdf = np.cumsum(pdf)
plt.plot(bins_edges[1:],pdf,color='r',label='pdf')
plt.plot(bins_edges[1:],cdf,color='g',label='cdf')
plt.xlabel('petal_length')
plt.ylabel('probability')
plt.legend()
plt.grid()
plt.show()
```

```python
count,bins_edges = np.histogram(iris['petal.width'],bins=10,density=True)
pdf = count/(sum(count))
print(pdf)
print(bins_edges)

```

```python
cdf = np.cumsum(pdf)
plt.plot(bins_edges[1:],pdf,color='r',label='pdf')
plt.plot(bins_edges[1:],cdf,color='g',label='cdf')
plt.xlabel('petal_width')
plt.ylabel('probability')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calculate mean, median , standard deviation of each type of species.
g = iris.groupby('variety')
for group_name,items in g:
    print(group_name)
    print(items)
```

```python
g.describe()
```

```python
# Calculate percentile of the data.
# A more specific term called quantiles of the data.
# 1st quantiles is 25%
# 2nd quntiles is called median(50%)
# 3rd and 4th quantiles are respectivly 75% and 100% .
for group_names,items in g:
    print()
    print("-------------------")
    print(group_names)
    print("0%\t25%\t50%\t75%")
    for col in items.columns[:4]:
        print(col)
        print(np.percentile(items[col],np.arange(0,100,25)))
```

```python
# Calculating 90th percentile
for group_name, items in g:
    print()
    print("-------------------")
    print(group_name)
    for col in items.columns[:4]:
        print(col)
        print(np.percentile(items[col],90))
```

```python
# Calculate mad
from statsmodels import robust
for group_name , items in g:
    print()
    print('-----------------------------')
    print(group_name)
    for col in items.columns[:4]:
        print(col)
        print(robust.mad(items[col]))
```

```python
# BoxPlot
col = iris.columns
print(col)
for i in col[:4]:
    sns.boxplot(x=col[4],y=i,data=iris)
    plt.show()
    plt.close()
```

```python
# violinPlot -> pdf+histogram + boxplot
for i in col[:4]:
    sns.violinplot(x=col[4],y=i,data=iris)
    plt.show()
    plt.close()
```
