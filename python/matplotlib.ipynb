```python
import matplotlib.pyplot as plt

yAxis= [2,4,6,3,5]

plt.plot(yAxis)

plt.xlabel("Indices")
plt.ylabel("Values")
plt.title("MyPlot")
plt.show()
```

```python
#Show grid views
import numpy as np
xAxis = np.arange(1,5,1)
yAxis = xAxis*xAxis
plt.plot(xAxis,yAxis)
plt.xlabel("Number")
plt.ylabel("Square")
plt.grid()
plt.title("Square of number")
plt.show()
```

```python
#Show grid views not the line view use dot views
xAxis = np.arange(1,5,1)
yAxis = xAxis*xAxis
plt.plot(xAxis,yAxis,'ro')
plt.xlabel("Number")
plt.ylabel("Square")
plt.grid()
plt.title("Square of number")
plt.show()

```

```python
#Draw graph with legend
xAxis = np.arange(0,5,0.2)
plt.plot(xAxis,xAxis,'b--',label='linear')
plt.plot(xAxis,xAxis**2,'gs',label='Square')
plt.plot(xAxis,xAxis**3,'r^',label='Cube')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Comparison")
plt.legend()
plt.show()
```

```python
# How to use the  set function
x1 = np.arange(0,5,1)
y1 = x1**2
x2 = np.arange(0,5,1)
y2 = x2**3
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color='r',linewidth = 2.0,label='Square')
plt.setp(lines[1],'color','g','linewidth','1.0','label','cube')
plt.grid()
plt.legend()
plt.show()
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

x1 = np.arange(2,7,0.4)
x2 = np.arange(1,4,0.3)
plt.figure(2)
plt.subplot(211)
plt.plot(x1,f(x1),label='first',color = 'r')
plt.grid()
plt.legend()
plt.plot()
plt.subplot(212)
plt.plot(x2,f(x2),color='b',label='second')
plt.legend()
plt.plot()
plt.show()


```
