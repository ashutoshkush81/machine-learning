# <font color = 'Green'>1. Uniform distribution (continuous) </font>


The probability density function of the continuous uniform distribution is:

${\displaystyle f(x)={\begin{cases}{\frac {1}{b-a}}&\mathrm {for} \ a\leq x\leq b,\\[8pt]0&\mathrm {for} \ x<a\ \mathrm {or} \ x>b\end{cases}}}$

The values of f(x) at the two boundaries a and b are usually unimportant because
they do not alter the values of the integrals of f(x) dx over any interval,
 nor of x f(x) dx or any higher moment.

 Notation	${\displaystyle:  {\mathcal {U}}(a,b)}$  or  ${\displaystyle \mathrm {unif} (a,b)} $

 Mean : ${\tfrac {1}{2}}(a+b)$

 Median : ${\tfrac {1}{2}}(a+b)$

 Variance : 	${\displaystyle {\tfrac {1}{12}}(b-a)^{2}}$

 Skewness : $0$

 Kurtosis : ${-\tfrac{6}{5}}$

```python
# generate the random integer between a and b by using the random.uniform(low,high,size) function in numpy.
from matplotlib import pylab

arr = np.random.uniform(low = 10, high = 20,size = 10)
print(arr)

print('mean : ' + str(np.mean(arr)))
print('varience : '+ str(np.std(arr)) )
```

Quantile:	${\displaystyle \mu +\sigma {\sqrt {2}}\operatorname {erf} ^{-1}(2p-1)}$

Mean:	${\displaystyle \mu } $

Median: 	${\displaystyle \mu }$

Mode:   	${\displaystyle \mu }$

Variance:	${\displaystyle \sigma ^{2}}$


MAD(Average absolute deviation):	${\displaystyle \sigma {\sqrt {2/\pi }}}$

# <font color = 'Green'> Box-Cox Transformation</font>

It's very useful tools to convert some distribution into normal distribution.

Most of time the data are not distributed according to the normal distribution.
Visualization of such data are very difficult. Hence we try to convert into normal distribution
to visualize data clearly.


```python
# Import train_test_split from sklearn library
from sklearn.model_selection import train_test_split
# generate non-normal

original_data = np.random.exponential(size = 1000)
# sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
sns.set_style('whitegrid')
ax = sns.distplot(original_data,bins = 1000,hist = True,kde = True,hist_kws={'cumulative':False})
ax.set(xlabel = 'Exponential Distribution',ylabel = 'Frequency')
plt.show()

```

As we can see the above distribution is not symmetric.


```python
# split into testing & training data
train,test = train_test_split(original_data, shuffle=False)

# transform training data & save lambda value
train_data,fitted_lambda = stats.boxcox(train)

# use lambda value to transform test data
test_data = stats.boxcox(test, fitted_lambda)

# (optional) plot train & test
fig, ax=plt.subplots(1,2)
sns.distplot(train_data, ax=ax[0])
sns.distplot(test_data, ax=ax[1])
```

As we can see now it's convert the same data into the symmetric distribution.

#  Finding Relationship between two or more variables

# 1. Co-variance

In probability theory and statistics, covariance is a measure of the joint
variability of two random variables. If the greater values of one
variable mainly correspond with the greater values of the other variable,
 and the same holds for the lesser values, the covariance is positive.

Covariance for discrete values:

 ${\displaystyle \operatorname {cov} (X,Y)={\frac {1}{n}}\sum _{i=1}^{n}(x_{i}-E(X))(y_{i}-E(Y)).}$

 Covariance for continues values:

${\displaystyle \operatorname {cov} (X,Y)=\operatorname {E} {{\big [}(X-\operatorname {E} [X])(Y-\operatorname {E} [Y]){\big ]}}}$


Note:
1. If $cov(X,Y) > 0$ then if(x increases y also increases) and vice versa.
2. If $cov(X,y) < 0$ then  it opposite of previous one.

# 2. Pearson correlation coefficient
Pearson's correlation coefficient when applied to a population is
commonly represented by the Greek letter $Ï$ (rho) and may be referred
to as the population correlation coefficient or the population Pearson
correlation coefficient. Given a pair of random variables
 ${\displaystyle (X,Y)}$, the formula for $Ï$ is:

${\displaystyle \rho _{X,Y}={\frac {\operatorname {cov} (X,Y)}{\sigma _{X}\sigma _{Y}}}}$

where:
${\displaystyle \operatorname {cov} }$ is the covariance.

${\displaystyle \sigma _{X}}$ is the standard deviation of ${\displaystyle X}$

${\displaystyle \sigma _{Y}}$  is the standard deviation of ${\displaystyle Y}$

# 3.Spearman's rank correlation coefficient

The Spearman correlation coefficient is defined as the
Pearson correlation coefficient between the rank variables.

For a sample of size n, the n raw scores ${\displaystyle X_{i},Y_{i}}$ are converted to
ranks ${\displaystyle \operatorname {rg} X_{i},\operatorname {rg} Y_{i}}$, and ${\displaystyle r_{s}}$is computed as

${\displaystyle r_{s}=\rho _{\operatorname {rg} _{X},\operatorname {rg} _{Y}}={\frac {\operatorname {cov} (\operatorname {rg} _{X},\operatorname {rg} _{Y})}{\sigma _{\operatorname {rg} _{X}}\sigma _{\operatorname {rg} _{Y}}}},}$

where

${\displaystyle \rho }$ denotes the usual Pearson correlation coefficient,
 but applied to the rank variables,
${\displaystyle \operatorname {cov} (\operatorname {rg} _{X},\operatorname {rg} _{Y})} $ is the covariance of the rank variables,
${\displaystyle \sigma _{\operatorname {rg} _{X}}}$ and ${\displaystyle \sigma _{\operatorname {rg} _{Y}}}$ are the standard deviations of the rank variables.

# <font color = 'Green'>Confidence interval(C.I.)</font>

In statistics, a confidence interval (CI) is a type of estimate computed
from the statistics of the observed data. This proposes a range of plausible
 values for an unknown parameter (for example, the mean). The interval has
 an associated confidence level that the true parameter is in the proposed
 range. Given observations ${\displaystyle x_{1},\ldots ,x_{n}} $and a confidence level ${\displaystyle \gamma }$
 , a valid confidence interval has a ${\displaystyle \gamma }$
 probability of containing the true underlying parameter.

 In simple words , we have given some sample of unknown population and we have
 to find the some important properties like mean,medium and other information
 with the given sample.

 One thing we can do it we just simply calculate
the mean and medium and says it will be approximately equal to population mean, and so on.

The other option is somehow we can fine the range of mean,
median with some percentage of confidence.

 ${\displaystyle Z=\left({\frac {{\bar {X}} }{\sigma /\surd n}}\right)}$ as $n â†’ âˆž$, is the standard normal distribution.

# <font color = 'Green'> Case 2 : When we don't know the ${\displaystyle \sigma}$ :</font>

The t distribution (aka, Studentâ€™s t-distribution) is a probability
distribution that is used to estimate population parameters when the
sample size is small and/or when the population variance is unknown.
 sample sizes are sometimes small, and often we do not know the standard deviation of the population. When either of these problems occur, statisticians rely on the distribution of the t statistic (also known as the t score), whose values are given by:

t = [ x - Î¼ ] / [ s / sqrt( n ) ]

where x is the sample mean, Î¼ is the population mean, s is the standard deviation of the sample, and n is the sample size. The distribution of the t statistic is called the t distribution or the Student t distribution.

# <font color = 'Green' > Empirical Bootstrap Based Confidence Interval</font>

Bootstrap is a powerful, computer-based method for statistical inference
 without relying on too many assumption.

 It is a re-sampling method by independently sampling with replacement
 from an existing sample data with same sample size n, and performing
 inference among these re-sampled data.

How to perform bootstrap technique to solve any problem ?
- We have n (size of sample data).
- We will make k sample from the given sample with size $m( <= n)$.
- Calculate mean , median or any statical number of each sample.
- Then we will lower and upper confidence of given data.

```python
import numpy as np
from sklearn.utils import resample
import matplotlib.pylab as plt
import seaborn as sns
# Sample data
data = [180 , 165 , 158 , 172 , 168 , 150 , 171 , 183 , 165 ,176 ]

# Size of data
n = int(len(list(data)))
# Take k = 10000
k = 10000

boot_mean =[]
for i in range(k):
    s = resample(data,n_samples=n)
    m = np.mean(s)
    boot_mean.append(m)

plt.hist(boot_mean)

plt.show()
plt.close()

sns.set_style('whitegrid')
ax = sns.distplot(boot_mean,bins = 1000,hist = True,kde = True)
ax.set(xlabel = 'Means of sample',ylabel = 'PDF')
plt.show()
plt.close()

# Now calculate the 95% confidence interval

low_value = np.percentile(boot_mean,2.5)
high_value = np.percentile(boot_mean,97.5)
print("The mean of the population in range of "+ str(low_value) + " , "+str(high_value)+" with 95% of C.I.")
```

```python
import numpy as np
from sklearn.utils import resample
import matplotlib.pylab as plt
import seaborn as sns
# Sample data
data = [180 , 165 , 158 , 172 , 168 , 150 , 171 , 183 , 165 ,176 ]

# Size of data
n = int(len(list(data)))
# Take k = 10000
k = 10000

boot_median =[]
for i in range(k):
    s = resample(data,n_samples=n)
    m = np.median(s)
    boot_median.append(m)

plt.hist(boot_median,density=True,cumulative=False,histtype='bar')

plt.show()
plt.close()

sns.reset_defaults()
sns.set_style('darkgrid')
ax = sns.kdeplot(boot_median)
ax.set(xlabel = 'Medians of sample',ylabel = 'PDF')
plt.show()

# Now calculate the 95% confidence interval

low_value = np.percentile(boot_median,2.5)
high_value = np.percentile(boot_median,97.5)
print("The median of the population in range of "+ str(low_value) + " , "+str(high_value)+" with 95% of C.I.")
```

# <font color='Green'>HYPOTHESIS TESTING</font>

A Hypothesis test is a statistical test that is used to determine
whether there is enough evidence in a sample of data to infer that
a certain condition is true for the entire population.

![](hypothesis.png)

A hypothesis test examines two opposing hypotheses about a population:
- the null hypothesis and
- the alternative hypothesis.

## Null hypothesis ($H_0$):
The null hypothesis states that a population parameter is equal to a
value. The null hypothesis is often an initial claim that researchers
specify using previous research or knowledge.

## Alternative Hypothesis ($H_1$):
The alternative hypothesis states that the population parameter is
different than the value of the population parameter in the null
hypothesis. The alternative hypothesis is what you might believe to
be true or hope to prove true.


Based on the sample data, the test determines whether to reject the
null hypothesis. You use a p-value, to make the determination.
If the p-value is less than or equal to the level of significance,
which is a cut-off point that you define, and then you can reject the
null hypothesis.
A common misconception is that statistical hypothesis tests are designed
to select the more likely of two hypotheses. Instead, a test will
remain with the null hypothesis until there is enough evidence (data)
to support the alternative hypothesis.

For more Information check:
https://medium.com/@thecodingcookie/hypothesis-testing-92b7270976de

## How to Calculate p-value?
To calculate p-value we will use permutation test.

## Permutation Test :
 We shuffle and mix everything together to get a big pool of data and
 compare this pool with what we had before shuffling it.

### <font color='red'> Algorithm to measure the significance of the difference of a metric between two variables </font>

- Choose your metric which we will call it ${\tau}$. This can be any distribution metric like mean, median, kurtosis, etc.
- Compute the ground truth absolute difference $Î”Ï„$ between ${\tau_1}$ and ${\tau_2}$ from your two variables:
  ${\Delta_\tau}=| \tau_1 - \tau_2|$
- Pool your variables into one single distribution.
- Randomly sample (also called bootstrapping) without replacement two
distributions with the size equal to the original distributions from
this pooled distribution to compute the absolute difference $RÎ”Ï„$ of
your metric between your two permuted samples:${R\Delta_\tau}=| R\tau_1 - R\tau_2|$
and repeat this p times.
- Finally, the proportion of permuted differences higher than your ground truth difference is your significance value.





#### Q: Perform a simple hypothesis testing with significance value is 5% .



```python
import numpy as np
import seaborn as sns
from sklearn.utils import resample
# Let's say we have sample of height of length 50 for sample 1 and sample 2.
sample1 = np.random.uniform(160,185,50)
sample2 = np.random.uniform(155,181,50)
# Apply Hypothesis testing on mean of two different sample.
# Step1 : Calculate mean - height of sample1 and sample2.
mean_1 = np.mean(sample1)
mean_2 = np.mean(sample2)
# H0 - There is no difference between the mean of height of two different population.
# H1 - There is a difference between the mean of height of two different population.
# Calculate Delta
Delta = abs(mean_1 - mean_2)
# Pool our value into single sample
Sample = np.concatenate([sample2,sample1])

mean_sample=[]
p = 10000
for i in range(p):
    s1 = resample(Sample,n_samples=50)
    s2 = resample(Sample,n_samples=50)
    mean_sample.append(abs(np.mean(s1)-np.mean(s2)))

# Now we have mean sample Firstly we will draw the distribution of mean_difference
sns.set_style('darkgrid')
ax=sns.distplot(mean_sample,bins=1000,hist=True,kde=True)
ax.set(xlabel = '$|{\mu_1}-{\mu_2}|$',ylabel = "PDF" )
ax.axvline(2.8, 0,0.25)
# Now calculate p-value.
p_val = len(np.where(mean_sample>=Delta)[0])/p
print(p_val)
```

P_value for above problem is nearly equal to 1.7 < 5.
So I will reject the null hypothesis.
So we can say that there is differnce between the mean of two populations.

# <font color='Green'> Kolmogorovâ€“Smirnov test(KS TEST) </font>

The KS test is a non-parametric and distribution-free test:
- It makes no assumption about the distribution of data.
- The KS test can be used to compare a sample with a reference probability distribution, or to compare two samples.

Suppose we have observations $x_1, x_2, â€¦x_n$ that we think come from a distribution P. The KS test is used to evaluate:
- Null Hypothesis: The samples do indeed come from P
- Alternative Hypothesis: The samples do not come from P

## Note:
The Kolmogorovâ€“Smirnov statistic for a given cumulative distribution function F(x) is

${\displaystyle D_{n}=\sup _{x}|F_{n}(x)-F(x)|}$

where $sup _{x}$ is the supremum of the set of distances.

${F(x)}$ is cumulative distribution of given sample.

The Kolmogorovâ€“Smirnov test may also be used to test whether two underlying one-dimensional probability distributions differ.
In this case, the Kolmogorovâ€“Smirnov statistic is

${\displaystyle D_{n,m}=\sup _{x}|F_{1,n}(x)-F_{2,m}(x)|,}$

where ${\displaystyle F_{1,n}}$ and ${\displaystyle F_{2,m}}$ are the empirical distribution functions of the first and the second sample respectively,
 and ${\displaystyle \sup }$ is the supremum function.

For large samples, the null hypothesis is rejected at level
${\displaystyle \alpha }$  if

${\displaystyle D_{n,m}>c(\alpha ){\sqrt {\frac {n+m}{n\cdot m}}}.}$

Where ${\displaystyle n}$ and ${\displaystyle m}$ are the sizes of first and second sample respectively.

The value of ${\displaystyle c({\alpha })}$ is given by :

${\displaystyle c\left(\alpha \right)={\sqrt {-\ln \left({\tfrac {\alpha }{2}}\right)\cdot {\tfrac {1}{2}}}},}$

```python
import numpy as np
import seaborn as sns
from scipy.stats import kstest
import matplotlib.pylab as plt

# Create an array with normal distribution of size 100.
arr = np.random.normal(0,1,100)
# Create another array with uniform distribution of size 200.
arr2 = np.random.uniform(0,1,200)
# Draw the distribution
sns.set_style('darkgrid')
sns.kdeplot(np.array(arr))
plt.show()
# plt.close()
sns.kdeplot(np.array(arr2))
plt.show()
plt.close()

print(kstest(arr,'norm'))
print(kstest(arr,'uniform'))
print(kstest(arr2,'norm'))
print(kstest(arr2,'uniform'))
```

As we can see the p-value for first is high so we can say that arr has std-normal distribution.
So we will accept the null hypothesis.
In second case we will reject the null hypothesis.
In third case we will again reject the null hypothesis.
In fourth case we will accept the null hypothesis.

In above example we uses the some standard distribution to check they are similar or not.
But how to complare two actual distributin which you don't know about.

# Proportional Sampling :
PPS sampling, in which the selection probability for each element is set to be proportional to its size measure, up to a maximum of 1.
There are total 4 steps involved to implement proportional sampling.
Let element are ${X_1 , X_2 , X_3 .... X_n}$ we have to implement proportional sampling such that the higher the value of ${X_i}$ higher will be the 
probability of selection.
Steps are following :
- First calculate to cumulative sum of all element.
- Normalize all the element by dividing every element by it's cumulative sum.
- Calculate cumulative normal value for every element(i,e. sum of all the normalized value above it).
- Now every value will be in between 0 to 1(in increasing order).
- Now use the uniform distribution to generate a single value in between 0 to 1.
- Using the generated value search the element which is greater than or equal to above value.
- Now return the original value that we get from the above procedure.




```python

import numpy as np
import matplotlib.pylab as plt
#-----------------------------
def upper_bound(nums, target):
    l, r = 0, len(nums) - 1
    while l <= r:
        mid = int(l + (r - l) / 2)
        if nums[mid] > target:
            r = mid - 1
        else:
            l = mid + 1
    return l
#----------------------------

arr = np.random.randint(10,40,10)
print(arr)

Total = np.sum(arr)
print(Total)

newarr = arr/Total
print(newarr)

newarr = np.cumsum(newarr)
print(newarr)

finalarr=[]
for i in newarr:
    finalarr.append(i)

for i in range(1000):
    s = np.random.uniform(0,1,1)
    idx = upper_bound(finalarr,s)
    obtainded_index.append(idx)
    

plt.hist(obtainded_index)
plt.show()
    
```
