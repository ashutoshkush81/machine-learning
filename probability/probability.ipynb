# <font color = 'Green'>1. Uniform distribution (continuous) </font>


The probability density function of the continuous uniform distribution is:

${\displaystyle f(x)={\begin{cases}{\frac {1}{b-a}}&\mathrm {for} \ a\leq x\leq b,\\[8pt]0&\mathrm {for} \ x<a\ \mathrm {or} \ x>b\end{cases}}}$

The values of f(x) at the two boundaries a and b are usually unimportant because
they do not alter the values of the integrals of f(x) dx over any interval,
 nor of x f(x) dx or any higher moment.

 Notation	${\displaystyle:  {\mathcal {U}}(a,b)}$  or  ${\displaystyle \mathrm {unif} (a,b)} $

 Mean : ${\tfrac {1}{2}}(a+b)$

 Median : ${\tfrac {1}{2}}(a+b)$

 Variance : 	${\displaystyle {\tfrac {1}{12}}(b-a)^{2}}$

 Skewness : $0$

 Kurtosis : ${-\tfrac{6}{5}}$

```python
# generate the random integer between a and b by using the random.uniform(low,high,size) function in numpy.
from matplotlib import pylab

arr = np.random.uniform(low = 10, high = 20,size = 10)
print(arr)

print('mean : ' + str(np.mean(arr)))
print('varience : '+ str(np.std(arr)) )
```

The <font color = 'red'>uniform function </font>generates a uniform continuous variable between the specified interval via its loc and scale arguments. This distribution is constant between loc and loc + scale. The size arguments describe the number of random variates. If you want to maintain reproducibility, include a random_state argument assigned to a number.

```python
from scipy.stats import uniform

n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)

sns.set_style('whitegrid')
ax = sns.distplot(data_uniform,bins = 1000,hist = False,kde = True)
ax.set(xlabel = 'Uniform Distribution',ylabel = 'Frequency')
plt.show()
```

${\displaystyle f(x)={\frac {1}{\sigma {\sqrt {2\pi }}}}e^{-{\frac {1}{2}}\left({\frac {x-\mu }{\sigma }}\right)^{2}}}$

            $ {\displaystyle \sigma ^{2}>0}\sigma ^{2}>0 = variance (squared scale)$
            
            
Support:   ${\displaystyle x\in \mathbb {R} }{\displaystyle x\in \mathbb {R} }$

PDF:	${\displaystyle {\frac {1}{\sigma {\sqrt {2\pi }}}}e^{-{\frac {1}{2}}\left({\frac {x-\mu }{\sigma }}\right)^{2}}}$

Quantile:	${\displaystyle \mu +\sigma {\sqrt {2}}\operatorname {erf} ^{-1}(2p-1)}$

Mean:	${\displaystyle \mu } $

Median: 	${\displaystyle \mu }$

Mode:   	${\displaystyle \mu }$

Variance:	${\displaystyle \sigma ^{2}}$


MAD(Average absolute deviation):	${\displaystyle \sigma {\sqrt {2/\pi }}}$

Skewness:	${\displaystyle 0}$

Let ${\displaystyle Z}$ be a standard normal variable, and let ${\displaystyle \mu }$  and ${\displaystyle \sigma >0}$ be two real numbers. Then, the distribution of the random variable

${\displaystyle X=e^{\mu +\sigma Z}}$

is called the log-normal distribution with parameters ${\displaystyle \mu }$ and ${\displaystyle \sigma }$ . Thus, these are the expected value (or mean) and standard deviation of the variable's natural logarithm, not the expectation and standard deviation of ${\displaystyle X}$ itself.

# <font color = 'Green'> Box-Cox Transformation</font>

It's very useful tools to convert some distribution into normal distribution.

Most of time the data are not distributed according to the normal distribution.
Visualization of such data are very difficult. Hence we try to convert into normal distribution
to visualize data clearly.


```python
# Import train_test_split from sklearn library
from sklearn.model_selection import train_test_split
# generate non-normal

original_data = np.random.exponential(size = 1000)
# sns.set_style("ticks", {"xtick.major.size": 8, "ytick.major.size": 8})
sns.set_style('whitegrid')
ax = sns.distplot(original_data,bins = 1000,hist = True,kde = True,hist_kws={'cumulative':False})
ax.set(xlabel = 'Exponential Distribution',ylabel = 'Frequency')
plt.show()

```

As we can see the above distribution is not symmetric.


```python
# split into testing & training data
train,test = train_test_split(original_data, shuffle=False)

# transform training data & save lambda value
train_data,fitted_lambda = stats.boxcox(train)

# use lambda value to transform test data
test_data = stats.boxcox(test, fitted_lambda)

# (optional) plot train & test
fig, ax=plt.subplots(1,2)
sns.distplot(train_data, ax=ax[0])
sns.distplot(test_data, ax=ax[1])
```

As we can see now it's convert the same data into the symmetric distribution.

#  Finding Relationship between two or more variables

# 1. Co-variance

In probability theory and statistics, covariance is a measure of the joint
variability of two random variables. If the greater values of one
variable mainly correspond with the greater values of the other variable,
 and the same holds for the lesser values, the covariance is positive.

Covariance for discrete values:

 ${\displaystyle \operatorname {cov} (X,Y)={\frac {1}{n}}\sum _{i=1}^{n}(x_{i}-E(X))(y_{i}-E(Y)).}$

 Covariance for continues values:

${\displaystyle \operatorname {cov} (X,Y)=\operatorname {E} {{\big [}(X-\operatorname {E} [X])(Y-\operatorname {E} [Y]){\big ]}}}$


Note:
1. If $cov(X,Y) > 0$ then if(x increases y also increases) and vice versa.
2. If $cov(X,y) < 0$ then  it opposite of previous one.

# 2. Pearson correlation coefficient
Pearson's correlation coefficient when applied to a population is
commonly represented by the Greek letter $Ï$ (rho) and may be referred
to as the population correlation coefficient or the population Pearson
correlation coefficient. Given a pair of random variables
 ${\displaystyle (X,Y)}$, the formula for $Ï$ is:

${\displaystyle \rho _{X,Y}={\frac {\operatorname {cov} (X,Y)}{\sigma _{X}\sigma _{Y}}}}$

where:
${\displaystyle \operatorname {cov} }$ is the covariance.

${\displaystyle \sigma _{X}}$ is the standard deviation of ${\displaystyle X}$

${\displaystyle \sigma _{Y}}$  is the standard deviation of ${\displaystyle Y}$

# 3.Spearman's rank correlation coefficient

The Spearman correlation coefficient is defined as the
Pearson correlation coefficient between the rank variables.

For a sample of size n, the n raw scores ${\displaystyle X_{i},Y_{i}}$ are converted to
ranks ${\displaystyle \operatorname {rg} X_{i},\operatorname {rg} Y_{i}}$, and ${\displaystyle r_{s}}$is computed as

${\displaystyle r_{s}=\rho _{\operatorname {rg} _{X},\operatorname {rg} _{Y}}={\frac {\operatorname {cov} (\operatorname {rg} _{X},\operatorname {rg} _{Y})}{\sigma _{\operatorname {rg} _{X}}\sigma _{\operatorname {rg} _{Y}}}},}$

where

${\displaystyle \rho }$ denotes the usual Pearson correlation coefficient,
 but applied to the rank variables,
${\displaystyle \operatorname {cov} (\operatorname {rg} _{X},\operatorname {rg} _{Y})} $ is the covariance of the rank variables,
${\displaystyle \sigma _{\operatorname {rg} _{X}}}$ and ${\displaystyle \sigma _{\operatorname {rg} _{Y}}}$ are the standard deviations of the rank variables.

# <font color = 'Green'>Confidence interval(C.I.)</font>

In statistics, a confidence interval (CI) is a type of estimate computed
from the statistics of the observed data. This proposes a range of plausible
 values for an unknown parameter (for example, the mean). The interval has
 an associated confidence level that the true parameter is in the proposed
 range. Given observations ${\displaystyle x_{1},\ldots ,x_{n}} $and a confidence level ${\displaystyle \gamma }$
 , a valid confidence interval has a ${\displaystyle \gamma }$
 probability of containing the true underlying parameter.

 In simple words , we have given some sample of unknown population and we have
 to find the some important properties like mean,medium and other information
 with the given sample.

 One thing we can do it we just simply calculate
the mean and medium and says it will be approximately equal to population mean, and so on.

The other option is somehow we can fine the range of mean,
median with some percentage of confidence.

# <font color = 'Green'> Case 2 : When we don't know the ${\displaystyle \sigma}$ :</font>

The t distribution (aka, Studentâ€™s t-distribution) is a probability
distribution that is used to estimate population parameters when the
sample size is small and/or when the population variance is unknown.
 sample sizes are sometimes small, and often we do not know the standard deviation of the population. When either of these problems occur, statisticians rely on the distribution of the t statistic (also known as the t score), whose values are given by:

t = [ x - Î¼ ] / [ s / sqrt( n ) ]

where x is the sample mean, Î¼ is the population mean, s is the standard deviation of the sample, and n is the sample size. The distribution of the t statistic is called the t distribution or the Student t distribution.

# <font color = 'Green' > Empirical Bootstrap Based Confidence Interval</font>

Bootstrap is a powerful, computer-based method for statistical inference
 without relying on too many assumption.

 It is a re-sampling method by independently sampling with replacement
 from an existing sample data with same sample size n, and performing
 inference among these re-sampled data.

How to perform bootstrap technique to solve any problem ?
- We have n (size of sample data).
- We will make k sample from the given sample with size $m( <= n)$.
- Calculate mean , median or any statical number of each sample.
- Then we will lower and upper confidence of given data.

```python
import numpy as np
from sklearn.utils import resample
import matplotlib.pylab as plt
import seaborn as sns
# Sample data
data = [180 , 165 , 158 , 172 , 168 , 150 , 171 , 183 , 165 ,176 ]

# Size of data
n = int(len(list(data)))
# Take k = 10000
k = 10000

boot_mean =[]
for i in range(k):
    s = resample(data,n_samples=n)
    m = np.mean(s)
    boot_mean.append(m)

plt.hist(boot_mean)

plt.show()
plt.close()

sns.set_style('whitegrid')
ax = sns.distplot(boot_mean,bins = 1000,hist = True,kde = True)
ax.set(xlabel = 'Means of sample',ylabel = 'PDF')
plt.show()
plt.close()

# Now calculate the 95% confidence interval

low_value = np.percentile(boot_mean,2.5)
high_value = np.percentile(boot_mean,97.5)
print("The mean of the population in range of "+ str(low_value) + " , "+str(high_value)+" with 95% of C.I.")
```

We can calculate other parameter in above way.
Let's try for median of population

```python
import numpy as np
from sklearn.utils import resample
import matplotlib.pylab as plt
import seaborn as sns
# Sample data
data = [180 , 165 , 158 , 172 , 168 , 150 , 171 , 183 , 165 ,176 ]

# Size of data
n = int(len(list(data)))
# Take k = 10000
k = 10000

boot_median =[]
for i in range(k):
    s = resample(data,n_samples=n)
    m = np.median(s)
    boot_median.append(m)

plt.hist(boot_median,density=True,cumulative=False,histtype='bar')

plt.show()
plt.close()

sns.reset_defaults()
sns.set_style('darkgrid')
ax = sns.distplot(boot_median,bins = 1000,hist = True,kde = True)
ax.set(xlabel = 'Medians of sample',ylabel = 'PDF')
plt.show()

# Now calculate the 95% confidence interval

low_value = np.percentile(boot_median,2.5)
high_value = np.percentile(boot_median,97.5)
print("The median of the population in range of "+ str(low_value) + " , "+str(high_value)+" with 95% of C.I.")
```
